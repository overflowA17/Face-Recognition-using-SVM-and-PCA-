```python
import os
import numpy as np

data_path = 'trainData'

labels = os.listdir(data_path)
categories = np.arange(len(labels))

categoriesDict = dict(zip(labels, categories))


print('labels: ',labels)
print('categories: ',categories)
print('categories dictionary: ',categoriesDict)
```


# to see the images

```python
import cv2

#face detection and cropping

face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

data=[]
target=[]

         
```

```python
print('Print y to accept image ')  

for label in labels:
    imgs_path = os.path.join(data_path, label)
    imgs_name = os.listdir(imgs_path) #gets image path
    
    for img_name in imgs_name:#goes through all the images
        img_path = os.path.join(imgs_path, img_name)
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray)
        
        for(x,y,w,h)in faces:
            #cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
            cropped_face = gray[y:y+h, x:x+w]
            cv2.imshow('Live', cropped_face)
            key = cv2.waitKey(0)
            #press y to accept image
            if(key == 121):
                
                cropped_face = cv2.resize(cropped_face, (50,50))
                data.append(cropped_face)
                target.append(categoriesDict[label])
                print('image is accepted!!! ')
            
            else:
                pass
cv2.destroyAllWindows()
```

```python
data = np.array(data)
target = np.array(target)

noImages, height, width = data.shape

data = data.reshape(noImages, height*width)

np.save('data',data)
np.save('target',target)
```
